Decision Tree is a very powerfull machine learning algorithm and forms the basis for couple of most popular machine learning algorithms i.e RandomForests and Boosted Trees. 
It is a supervised machine learning algorithm, used for both classification and regression task. 
The intuition is based on how humans tend to classify things through a set of rules.



It provides wonderful transparency for human interpretation by graphically representing the process of decision-making. Using this, we can predict both categorical variable and continuous variable.

It does this by creating a flow chart like structure, and an if-else condition is applied at the nodes on the attributes (if male or female?). From here the condition outcome is represented by branches and class label is represented by the leaf node. So, a rule is created at the decision nodes and a result is delivered at the leaf node. The entire flow is from root to leaf.


Types of trees
ID3 (Iterative Dichotomiser 3)
CART (Classification and Regression Tree)
CHAID (Chi-squared Automatic Interaction Detector)
C4.5 (Extension of ID3)


General Flow:
1. Start at the root node
2. Split data into groups (based on some criteria)
3. Set a decision at node
4. Move the data along the respective branches
5. Repeat the process untill a stopping criterion is met (max levels/depth, min samples left to split, nothing left to split etc)


Choosing the root node:
Regression:
A splitting point is choosen such that there is the greatest reduction in RSS (Resudual Sum of Squares)


\begin{equation*}
\sum (y_i - \hat{y_i}R_1)^2 + \sum (y_i -\hat{y_i}R_2)^2
\end{equation*}


Here YR1 and YR2 are mean responses of region 1 and 2. Once we train the tree, we predict the response for a test data using the mean of the training observations in that group.
 
Classification:
ID3 uses Entropy and Information Gain. 
CART uses Gini Index 


Entropy

Entropy is amount of information needed to accurately describe the sample data. So if sample is homogeneous, means all the data points are similar then Entropy is 0 while if the data is ransom then entropy is maximum (1)

Entropy quantifies the uncertainity of chaos in the group. Higher entropy means higher the disorder. It is denoted by H(x), where x is a vectir with probabilities p1, p2, p3....


\begin{equation*}
{Entropy}= - \sum_{i=1}^n p_i * log_2(p_i)
\end{equation*}



From the figure, we can see that the entropy (uncertainity) is highest when the probability is 0.5 i.e 50-50 chances. And entoropy is lowest when probability is 0 or 1 i.e there is no uncertainity in the system

So entropy is maximum if in a class there are equal number of objects from different attributes (like the goup has 50 cats and 50 dogs), and this is minimum if the node is pure (like the group has only 100 cats or 100 dogs). We untimately want to have minimum entrioy for the tree i.e pure or uniform classes at the leaf nodes.

**The more you know about a topic, the less new information you are apt to get about it. To be more concise: If you know an event is very probable, it is no surprise when it happens, that is, it gives you little information that it actually happened.**

In classification tree, while splitting the tree we select those attributes that achieves the greatest reduction in entropy. This reduction (or change) in entropy is meassured by **Information Gain**

\begin{equation*}
{Entropy(S)}= - \sum_{i=1}^n p_i * log_2(p_i) \\      
{Information Gain (S,A)} = {Entropy(S)} - \sum P(S|A) * Entropy(S|A)
\end{equation*}


We consider some toy problem. The problem is about predicting whether some kid is going to eat a particular type of food given that kids prior eating habits.



**Solution**
Entropy of the total attributes at the root node

No of 'No' -> 4          
No of 'Yes' -> 6          
Total Count -> 10

$\begin{aligned}
E_o  &= \sum_{i=1}^2[-P_i log_2(P_i)] \\ 
     &= -\frac{4}{10} log_2(\frac{4}{10}) -\frac{6}{10} log_2(\frac{6}{10}) \\

\end{aligned}$

$\begin{aligned}

IG_{Taste} &= E_o - E_{Taste}\\

Lets now look at GINI index        

Gini index is measure of inequality in data. It has value between 0 and 1.
A Gini score gives an idea of how good a split is by how mixed the classes are in the two groups created by the split. A perfect seperation results in a GINI score of 0, whereas the worst case split that results in 50/50 classes in each group results in a GINI score of 1.0 (for a 2 class problem)

$\begin{aligned}
Gini(t) = 1- \sum_{i=0}^{c-1}[p(i|t)]^2
\end{aligned}$
